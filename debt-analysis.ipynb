```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel('subnational-debt-OECD.xlsx', sheetname="Sheet1")
```

```python
gdpPerKm = np.arange(0,31)
count = 0
for i in df["GDP"]:
    gdpPerKm[count] = (i * 1000000000) / df["Area"][count]
    count = count + 1

    
df["GDP per km"] = gdpPerKm.tolist()
```

```python
ax = sns.jointplot(x=df["Debt per km squared"], y=df["GDP per km"], kind="reg")
ax.text(df["Debt per km squared"][3], df["GDP per km"][3], color="red")
```
