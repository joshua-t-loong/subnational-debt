
# 3D scatterplot


Demonstration of a basic scatterplot in 3D.



```python
import numpy as np
import pandas as pd

df = pd.read_excel('subnational-debt-OECD.xlsx', sheetname="Sheet1")

xs = df["Area"]
ys = df["Debt per km squared"]
zs = df["Share of GDP"]
labels = df["Country"]

list = df[df["Area"] >= 3989887.5]
list
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(xs, ys, zs)
ax.scatter(xs.mean(), ys.mean(), zs.mean())

ax.text(xs[3], ys[3], zs[3], "CAN", color="r")
ax.text(xs[0], ys[0], zs[0], "AUS", color="r")
ax.text(xs[33], ys[33], zs[33], "USA", color="r")
    
ax.set_xlabel('Area')
ax.set_ylabel('Debt per km squared')
ax.set_zlabel('Share of GDP')



plt.show()
```
